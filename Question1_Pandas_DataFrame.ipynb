```python
# Pandas Dataframe Tasks

# Step 1: Create the Dataframe

# We need to define the data and the use pd for pandas Dataframe

import pandas as pd

data = {
    'StudentID': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    'Name': ['Alice','Bob','Charlie','Diana','Ethan','Fiona','George','Hannah','Ivan','Julia'],
    'Age': [20, 21, 19, 22, 20, 21, 23, 19, 22, 20],
    'Marks': [88, 75, 92, 65, 78, 85, 70, 95, 60, 80],
    'Grade': ['A','B','A','C','B','A','B','A','C','B']
}
df = pd.DataFrame(data)
print(df)
```

```python
# Step 2: Save this as CSV

df.to_csv('students.csv', index=False)
print("File saved as 'students.csv'.\n")
```

```python
# Step 3: Reload the CSV

new_df = pd.read_csv('students.csv')
print(new_df)
```

```python
# Step 4: Display the first 3 rows

new_df = pd.read_csv('students.csv')
print(new_df.head(3))
```

```python
# Step 5: Display statistical summary

new_df = pd.read_csv('students.csv')
print("\n------ Statistical summary-------")
print(new_df.describe())
```
