```python
import pandas as pd
import numpy as np

data = {
    'patient_id': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110,
                   111, 112, 113, 114, 115, 101, 107, 118, 119, 120],
    'age': ['25', '34', None, '45', '29', None, '38', '52', '27', '41',
            '33', 'unknown', '48', '26', '35', '25', '38', '31', None, '44'],
    'weight': ['70', '65', '80', None, '75', None, '68', '90', '72', '85',
               '78', None, '82', '69', 'N/A', '70', '68', '74', None, '88'],
    'blood_pressure': [120, 130, None, 140, 125, None, 135, None, 118, 145,
                      128, None, 138, 122, None, 120, 135, 126, None, 142],
    'medication': ['Aspirin', 'Metformin', 'Lisinopril', None, 'Aspirin',
                   'Metformin', 'Lisinopril', 'Aspirin', None, 'Metformin',
                   'Lisinopril', 'Aspirin', None, 'Metformin', 'Aspirin',
                   'Aspirin', 'Lisinopril', 'Metformin', 'Aspirin', None],
    'insurance_provider': ['Blue Cross', 'Aetna', 'Cigna', 'UnitedHealth', None,
                          'Blue Cross', 'Aetna', 'Cigna', 'UnitedHealth', 'Blue Cross',
                          'Aetna', None, 'UnitedHealth', 'Blue Cross', 'Aetna',
                          'Blue Cross', 'Aetna', 'Cigna', 'UnitedHealth', None]
}

df = pd.DataFrame(data)

# Task 1: Inspect the Data

# Step 1: Run the code and Interpret the output

print(df.info())
```

```python
# Step 2: To count the missing values per column

print(df.isnull().sum())
```

```python
# Step 3: Calculate the percentage of missing values

print((df.isnull().sum() / len(df)) * 100)
```

```python
# Step 4: To count the duplicate rows

print(df.duplicated().sum())
```

```python
# Task 2: Data Type Conversion

# Step 1: Convert age column to numeric

df['age'] = pd.to_numeric(df['age'], errors='coerce')

print(df['age'])
```

```python
# Step 2: Convert weight column to numeric

df['weight'] = pd.to_numeric(df['weight'], errors='coerce')

print(df['weight'])
```

```python
# Step 3: Check for new missing values

print(df.isnull().sum())
```

```python
# Step 4: Convert insurance_provider to category type

df['insurance_provider'] = df['insurance_provider'].astype('category')

print(df['insurance_provider'])
```

```python
# Step 5: Verify all the conversions

df.dtypes
```

```python
# Task 3: Handle Missing Values

# Step 1: Fill the missing age values with the median

df['age'] = df['age'].fillna(df['age'].median())

print(df['age'])
```

```python
# Step 2: Fill the missing weight values with the median

df['weight'] = df['weight'].fillna(df['weight'].median())
print(df['weight'])
```

```python
# Step 3: Fill the missing bloood_pressure values with the median

df['blood_pressure'] = df['blood_pressure'].fillna(df['blood_pressure'].median())
print(df['blood_pressure'])
```

```python
# Step 4: Fill missing medication values with the mode

df['medication'] = df['medication'].fillna(df['medication'].mode()[0])
print(df['medication'])
```

```python
# Step 5: Fill missing insurance_provider values with 'Unknown'

df['insurance_provider'] = df['insurance_provider'].cat.add_categories('Unknown')
df['insurance_provider'] = df['insurance_provider'].fillna('Unknown')
print(df['insurance_provider'])
```

```python
# Step 6: Verify no missing values

df.isnull().sum()
```

```python
# Task 4: Handle Duplicates

# Step 1: View duplicate rows

duplicates_all = df[df.duplicated(keep=False)]
print("All Duplicate Rows\n", duplicates_all)
```

```python
# Step 2: Identify duplicates on 'patient_id' only

duplicates_patient = df[df.duplicated(subset = ['patient_id'])]
print("\nDuplicates based on patient_id:\n", duplicates_patient)
```

```python
# Step 3: Remove duplicates keeping the first occurrence

shape_before = df.shape
df_clean = df.drop_duplicates(subset = ['patient_id'], keep='first')
shape_after = df_clean.shape
```

```python
# Step 4: Print shape before and after to show how many rows were removed
print(f"\nRows before: {shape_before[0]}")
print(f"Rows after: {shape_after[0]}")
print(f"Total rows removed: {shape_before[0] - shape_after[0]}")
```
