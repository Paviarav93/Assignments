```python
# Task 1: Print Shape & mean temperature per station

# Step 1: Defining the data array

import numpy as np

data = np.array([[22.5, 19.0, 31.2, 28.7, 25.1],
                 [17.3, 22.8, 30.5, 26.4, 21.9],
                 [33.1, 29.6, 18.4, 24.0, 27.8],
                 [20.2, 23.5, 31.9, 28.1, 22.6]])

# Step 2: Print the shape of the data

print("Shape of the data array:", data.shape)

# Step 3: Compute and print the mean temperature per station

mean_per_station = np.mean(data, axis=1)
print("Mean temperature per station:", mean_per_station)
```

```python
# Task 2: Extract temperatures above 28.0°C

# Step 1: Use a boolean mask to filter temperatures

threshold = 28.0
boolean_mask = data > threshold

# Step 2: Extracting the values and print as 1D array

extracted_temperatures = data[boolean_mask]
print("Temperature above 28.0°C:", extracted_temperatures)
```

```python
# Task 3: Normalize the data to range

# Step 1: Calculate the min and max values of the entire array

data_min = np.min(data)
data_max = np.max(data)

# Step 2: Apply normalization formula given in problem

normalized_data = (data - data.min()) / (data.max() - data.min())

# Step 3: Round the result to 2 decimal places & print

rounded_normalized_data = np.round(normalized_data, 2)
print("Normalized data(rounded to 2 decimal places):")
print(rounded_normalized_data)
```
